```python
import matplotlib.pyplot as plt
```

## **Step 2: Simple Line Plot**

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)   # simple line plot
plt.show() # This draws a line connecting the points.
```

## **Step 3: Adding Titles & Labels**

```python
plt.plot(x, y, color="blue", linewidth=2, marker="o")
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

## **Step 4: Multiple Lines in One Plot**

```python
y2 = [1, 4, 9, 16, 25]

plt.plot(x, y, label="2x", marker="o")
plt.plot(x, y2, label="x^2", marker="s")
plt.title("Multiple Lines")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.legend()
plt.show()
```

## **Step 5: Scatter Plot**

```python
x = [5, 7, 8, 7, 6, 9, 5, 6, 7, 8]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]

plt.scatter(x, y, color="red", marker="x")
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

## **Step 6: Bar Chart**

```python
# Categories
categories = ["A", "B", "C", "D"]
values = [3, 7, 5, 2]

# Bar chart
plt.bar(categories, values)

plt.title("Simple Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
# Custom Colors & Width
plt.bar(categories, values, color="orange", width=0.5)

plt.title("Bar Chart with Custom Color")
plt.show()
```

```python
# Horizontal Bar Chart
plt.barh(categories, values, color="green")

plt.title("Horizontal Bar Chart")
plt.xlabel("Values")
plt.ylabel("Categories")
plt.show()
```

```python
# Multiple Bar Charts (Comparison)
import numpy as np

categories = ["A", "B", "C", "D"]
values1 = [3, 7, 5, 2]
values2 = [4, 6, 8, 3]

x = np.arange(len(categories))   # positions [0,1,2,3], gives positions for bars.
width = 0.35

plt.bar(x - width/2, values1, width, label="Group 1")
plt.bar(x + width/2, values2, width, label="Group 2")

plt.xticks(x, categories)  # replace numbers with category names
plt.title("Grouped Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.legend()
plt.show()
```

## **Step 7: Histograms**

```python
# A histogram shows how data is distributed across intervals (called bins).
# Sample data (ages of people)
ages = [22, 25, 29, 30, 22, 40, 41, 38, 29, 30, 25, 32, 35, 36, 38, 40, 22, 25, 28, 35]

plt.hist(ages, bins=5, color="skyblue", edgecolor="black") # bins=5 divides ages into 5 intervals.

plt.title("Histogram of Ages")
plt.xlabel("Age Groups")
plt.ylabel("Frequency")
plt.show()
```

```python
# More Bins = More Detail
plt.hist(ages, bins=10, color="orange", edgecolor="black")

plt.title("Histogram with More Bins")
plt.show()
```

```python
# Normal Distribution Example

# Generate 1000 random values ~ Normal Distribution (mean=50, std=10)
data = np.random.normal(50, 10, 1000)

plt.hist(data, bins=20, color="green", edgecolor="black")

plt.title("Histogram of Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

## **Step 8: Pie Charts**

```python
# Data
labels = ["Apples", "Bananas", "Cherries", "Dates"]
sizes = [30, 10, 10, 50]

plt.pie(sizes, labels=labels)
plt.title("Fruit Distribution")
plt.show()
```

```python
plt.pie(sizes, labels=labels, autopct="%1.1f%%")
plt.title("Fruit Distribution with Percentages")
plt.show()
```

```python
# Highlight (Explode a Slice)
explode = [0.1, 0, 0, 0]  # "Apples" slice pops out

plt.pie(sizes, labels=labels, autopct="%1.1f%%", explode=explode, shadow=True, startangle=90)
plt.title("Pie Chart with Exploded Slice")
plt.show()
```

## **Step 9: Subplots**

```python
# Sometimes you need to show multiple plots in one figure for comparison.
# First plot (top-left)
# Using subplot() (manual control)
plt.subplot(2, 2, 1)   # (rows, cols, index)
plt.plot([1, 2, 3], [1, 4, 9])
plt.title("Plot 1")

# Second plot (top-right)
plt.subplot(2, 2, 2)
plt.bar([1, 2, 3], [3, 6, 9])
plt.title("Plot 2")

# Third plot (bottom-left)
plt.subplot(2, 2, 3)
plt.hist([1, 2, 2, 3, 3, 3, 4, 4, 4, 4])
plt.title("Plot 3")

# Fourth plot (bottom-right)
plt.subplot(2, 2, 4)
plt.pie([10, 20, 30], labels=["A", "B", "C"])
plt.title("Plot 4")

plt.tight_layout()  # Avoid overlapping
plt.show()
```

```python
# Using subplots() (recommended, cleaner)
fig, axes = plt.subplots(1, 2, figsize=(10, 4))  # 1 row, 2 columns, 1 row, 2 plots side by side.

# Left plot
axes[0].plot([1, 2, 3], [2, 4, 6])
axes[0].set_title("Line Plot")

# Right plot
axes[1].bar([1, 2, 3], [3, 6, 9])
axes[1].set_title("Bar Chart")

plt.tight_layout()
plt.show()
```

```python
# Looping through Subplots
fig, axes = plt.subplots(2, 2, figsize=(8, 6))

data = [ [1,2,3], [3,6,9], [2,4,6], [4,8,12] ]
titles = ["Plot A", "Plot B", "Plot C", "Plot D"]

for i, ax in enumerate(axes.flatten()): # flattens 2D grid into 1D array for easy looping.
    ax.plot(data[i])
    ax.set_title(titles[i])

plt.tight_layout() # automatically generating multiple plots.
plt.show()
```

## **Step 10: Styling & Customization in Matplotlib**

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 4, 9, 16, 25]

plt.plot(x, y1, color="red", linestyle="--", marker="o", label="y = 2x")
plt.plot(x, y2, color="blue", linestyle="-.", marker="s", label="y = x²")

plt.title("Styling Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()  # Show labels
plt.grid(True)  # Add grid

plt.show()
```

```python
plt.plot(x, y2, color="green", marker="^")

plt.title("Quadratic Growth", fontsize=16, fontweight="bold")
plt.xlabel("Input (X)", fontsize=12, color="purple")
plt.ylabel("Output (Y)", fontsize=12, color="purple")

plt.show()
```

```python
# style Sheets (Themes)
plt.style.use("seaborn-v0_8")  # or "ggplot", "fivethirtyeight", etc.

plt.plot(x, y1, label="Linear")
plt.plot(x, y2, label="Quadratic")

plt.legend()
plt.show()
```

## **Step 11: Saving Plots in Matplotlib**

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.plot(x, y, color="red", marker="o", label="Data")
plt.title("Saving Plot Example")
plt.legend()

plt.savefig("plot.png")  # Save as PNG
plt.show()
```

```python
# Save with Higher Resolution
plt.plot(x, y, color="blue", marker="s", label="Better Quality")
plt.legend()

plt.savefig("high_res_plot.png", dpi=300)  # 300 dpi = publication quality
plt.show()
```

```python
plt.plot(x, y, color="green", marker="^")
plt.title("Different Formats")

plt.savefig("plot.pdf")   # Save as PDF
plt.savefig("plot.jpg")   # Save as JPG
plt.savefig("plot.svg")   # Save as SVG (scalable)
plt.show()
```
